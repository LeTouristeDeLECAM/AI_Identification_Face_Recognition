# Crating a balanced dataset

```python
import os
import shutil
import random

eyeGlasses = 0
noEyeGlasses = 0
```

```python
# recurrent method to check if the dataset is balanced

def check_balance_2 (dataSetVolume,dataSetVolumeTrig, eyeGlasses, noEyeGlasses, linesOrg, linesCopy, interval ):

    try:
        ratio = eyeGlasses / noEyeGlasses
    except ZeroDivisionError:
        ratio = 0
        
    

    if dataSetVolume< dataSetVolumeTrig :
        

        # select random line with eye glasses and delete it
        random_line = random.choice(linesOrg)
        linesOrg.remove(random_line)

        # Count the balancement of the dataset
        if random_line.split()[1] == '1':
            eyeGlasses += 1
        else:
            noEyeGlasses += 1

        # Add to the list of the lines to be copied
        linesCopy.append(random_line)

        # Recurrent call
        check_balance_2(len(linesCopy),dataSetVolumeTrig, eyeGlasses, noEyeGlasses, linesOrg, linesCopy,interval )
    
    elif not ((0.5-(1-interval)) <= ratio <= (0.5+(1-interval))) :
        print("The dataset is not balanced")
        print ("Ratio: ", ratio)
        print("(0.5-(1-interval))",(0.5-(1-interval)),"(0.5+(1-interval))",(0.5+(1-interval)))


        # Remove in linesCopy of the excess data
        if ratio > (0.5+(1-interval)):
            # Remove eye glasses   ATTENTION 
            for line in linesCopy:
                if line.split()[1] == '1':
                    linesCopy.remove(line)
                    eyeGlasses -= 1
                    break



        else:
            # Remove no eye glasses ATTENTION 
            for line in linesCopy:
                if line.split()[1] == '0':
                    linesCopy.remove(line)
                    noEyeGlasses -= 1
                    break

        # Recurrent call
        check_balance_2(len(linesCopy),dataSetVolumeTrig, eyeGlasses, noEyeGlasses, linesOrg, linesCopy,interval )
        
        
    else : 
        print("The dataset is balanced")
        print("The ratio of eye glasses to no eye glasses is: ", ratio)
        print ("eyeGlasses: ", eyeGlasses, "noEyeGlasses: ", noEyeGlasses)
        print ("The number of lines in the linesCopy file is: ", len(linesCopy))
        print ("The number of lines in the linesOrg file is: ", len(linesOrg))
        print ("linesCopy: ", linesCopy)

        return linesCopy
```

```python
file_path_meta_data_org= "meta.txt"

# copy the original file to a new file
shutil.copyfile(file_path_meta_data_org, 'metaCopy.txt')

file_path_meta_data= "metaCopy.txt"

# Open the file and read its lines
with open(file_path_meta_data, 'r+') as f:
    linesOrg = f.readlines()

dataSetVolume=0
dataSetVolumeTrig=1250
eyeGlasses=0
noEyeGlasses=0
linesCopy=[]

check_balance_2 (dataSetVolume,dataSetVolumeTrig, eyeGlasses, noEyeGlasses, linesOrg, linesCopy , 0.90)
```
